```python
import os
import pandas as pd
import numpy as np
```

```python
path = './DataSet/'
#Getting the file list
files = []
# r=root, d=directories, f = files
for r, d, f in os.walk(path):
    for file in f:
        if '.csv' in file:
            files.append(os.path.join(r, file))

#289
dataset = pd.read_csv(files[0])
dataset.head()
```

## DS for clustering and classification
The goal of this block is to reshape data for analysis and mining models. The structure will be a dataset with 24 hours of data arranged in one row such marked agains the dataset/area it is extracted. The analysis will be done to recognize the types of areas on the basis of their daily traffic pattern.<br><br>
<b> Structure: <b>
- The rows will contain the areas their traffic details.
- The column will have the first 24 hours for the first date in the dataset

```python
new_data = []
#Adding new column tags
new_data.append(["Area"] + list(dataset.iloc[:288,0]))
c = 1
#Iterating data through files
for x in files:
    dataset = pd.read_csv(x)
    new_data.append(["A"+str(c)] + list(dataset.iloc[:288,1]))
    c+=1
```

```python
#Adding vertical rows
data = np.array(new_data)
dataframe = pd.DataFrame(data[1:],columns=data[0])
dataframe.head()
```

```python
#Converting Dataframe to CSV
dataframe.to_csv("Dataset.csv")
```

### Visualization Data
This data is to be used in visualization tools to ad-hoc analyse the data. This dataset will be a transpose of the previous version with the addition of day, month and hour.

```python
import datetime
```

```python
new_data = []
#Adding new column tags
new_data.append(["Area"] + list(dataset.iloc[:,0]))
c = 1
#Iterating data through files
for x in files:
    dataset = pd.read_csv(x)
    new_data.append(["A"+str(c)] + list(dataset.iloc[:,1]))
    c+=1
```

```python
vis = np.array(new_data).transpose()
```

```python
vis[0][0] = 'Date'
```

```python
day_of_week = []
month = []
hour = []
for i in range(1,len(vis)):
    #Getting the day of week for detecting weekends
    day = datetime.datetime.strptime(vis[i][0], '%Y-%m-%d %H:%M:%S').strftime('%a')
    day_of_week.append(day)  
    #Getting the month for vacation times
    mon = int(datetime.datetime.strptime(vis[i][0], '%Y-%m-%d %H:%M:%S').strftime('%m'))
    month.append(mon)
    hr = int(datetime.datetime.strptime(vis[i][0], '%Y-%m-%d %H:%M:%S').strftime('%H'))
    hour.append(hr)
    
    
```

```python
#Creating the dataframe
dataset = pd.DataFrame(vis[1:,:],columns=vis[0])
dataset.head()
```

```python
#Adding the new columns of day, month and hour
dataset['day'] = day_of_week
dataset['month'] = month
dataset['hour'] = hour
```

```python
dataset.head()
```

```python
#Saving the dataset
dataset.to_csv("Visualization.csv")
```
